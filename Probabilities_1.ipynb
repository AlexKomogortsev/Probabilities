```python
# Task 1

from math import factorial

def arrangements(n, k):
    return int(factorial(n) / factorial(n-k))

arrangements(10, 4)

# четыре предмета из 10ти можно расставить 5040 способами
```

```python
# Task 2

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n-k)))

# Найдём общее количество комбинаций, которыми можно вытащить 4 карты

a = combinations(52, 4)

# Найдём количество комбинаций, когда мы НЕ вытащим ни одного туза

b = combinations(4, 0) * combinations(48, 4)

# Получим количетсво комбинаций, когда есть хотя бы один туз

c = a - b

print(c)
```

```python
# Task 3

# Посадить одного человека с соседом справа на 7-ми местах можно 6-ю способами - для пары 1-2
# Посадить одного человека с соседом справа на 7-ми местах можно 6-ю способами - для пары 2-1
# Соответственно двух человек можно посадить 12-способами на 7-ми местах

# Общее количество размещений 7-ми человек на 7-ми местах равно:

def permutations(n):
    return int(factorial(n))

permutations(7)

# вероятность сесть рядом двум определённым людям

a = 12 / permutations(7)

print(a)

# вероятность меньше 1%
```

```python
# Task 4.1

# Вероятность того, что студент вытянет первый билет нужный

a1 = 50 / 60

# Вероятность того, что студент вытянет второй билет нужный

a2 = 49 / 59

# Вероятность того, что студент вытянет третий билет нужный

a3 = 48 / 58

# Общая вероятность вытянуть все три нужных билета

a = a1*a2*a3

print(a)
```

```python
# Task 4.2

# Вероятность того, что студент вытянет первый билет нужный

a1 = 50 / 60

# Вероятность того, что студент вытянет второй билет нужный

a2 = 49 / 59

# Вероятность того, что студент вытянет третий билет НЕ нужный

a3 = 10 / 58

# Вероятность вытянуть первые билета нужный, а последний НЕ нужный

a = a1*a2*a3

# Также мы должны рассмотреть варианты: когда первый билет плохой, а два остальных хорошие; 
# и вариант, когда второй билет плохой, остальные хорошие
# Учитвая, что общий результат является произведением величин a1, a2, a3, то он будет одинаковым для всех трёх вариантов
# Следовательно вероятность вытащить два хороших билета равна:

print(3*a)
```

```python
# Task 5

# Событие А: выпадание чисел 2, 4, 6
# Событие B: выпадание чисел 4, 5, 6

# При условии выполнения события B событие А может выпасть 2 из трёх раз (числа 4 и 6)
PAfromB = 2 / 3

# Вероятность выполнения собатия B равна 1 / 2 (3 числа из 6)
PB = 1 / 2

# Вероятность выпадения числа, удовлетворяющего обоим событиям
PAB = PAfromB * PB

# Вероятность выполнения собатия A равна 1 / 2 (3 числа из 6)
PA = 1 / 2

if PAB == PA:
    print('События А и В независимые')
else:
    print('События А и В зависимые')
```

```python
# Task 6

# Существует событие А, которое означает, что тест показал положительный результат. Также существуют события В1 и В2, когда
# человек болен, либо не болен соответственно. 


# Вероятность P(B1) того, что человек может заболеть равна 0,01 %

PB1 = 1 / 1000

# P(B2) - человек может быть здоровым

PB2 = 999 / 1000

# Вероятность того, что тест положительный при условии, что человек болен равна 99% - P(A|B1)

PAB1 = 99 / 100

# P(A|B2) - тест даёт положительный результат при условии, что человек здоров

PAB2 = 1 / 100

# Вероятность того, что тест в принципе может показать положительный результат равна P(A)

PA = PB1 * PAB1 + PB2 * PAB2

# И находим вероятность того, что человек болен при условии положительного теста - P(B1/A)

PBA1 = (PB1 * PAB1) / PA

print(PBA1)

```
